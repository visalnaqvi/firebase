```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_total_private_wealth"
response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
table = soup.find_all("table")
th = table[0].find_all("th")
headers = [header.get_text(strip=True) for header in th]
print(headers) 
```

```python
tbody = table[0].find("tbody")
rows = tbody.find_all("tr")
print(rows)
```

```python
data = []
for row in rows[4:]:  
    cells = row.find_all(["td", "th"])
    row_data = [cell.get_text(strip=True).replace('\u202f', ' ').replace('\u2014', '').replace('\xa0', ' ') for cell in cells]
    if row_data: 
        data.append(row_data)


df = pd.DataFrame(data, columns=headers[:len(data[0])])

df.to_csv("wealth_data.csv", index=False)

print(df.head())
```
